```python
#import libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("1000_Companies.csv")
X= df.iloc[:, :-1].values
y = df.iloc[:,4].values
```

```python
#first five rows
df.head()
```

```python
#encoding categorical data
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
column_transformer = ColumnTransformer([('encoder', OneHotEncoder(), [3])], remainder='passthrough')
X = column_transformer.fit_transform(X)
print(X)
```

```python
#Avoiding dunny variable trap
X=X[:, 1:]
```

```python
#Splitting the data into train and test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =  train_test_split(X,y, test_size = 0.2, random_state = 0)
```

```python
#Fitting the multiple linear regression dataset
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
#Predictin Test results
y_pred = regressor.predict(X_test)
y_pred
```

```python
#calculating the coefficients
print(regressor.coef_)
```

```python
#calculating the intercepts
print(regressor.intercept_)
```

```python
#calculating the R squared value
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
