# DECISION TREE

Decison Tree is a flow chart, and helps to make decisions based on the previous experience. In the example a person tries to decide whether to go to a comdey show or not.

```python
import pandas as pd

df = pd.read_csv (r"C:\Users\gojef\OneDrive\Desktop\data.csv")
df
```

Decision trees requires all data to be numerical. 

```python
#Convert non-numerical columns  'Nationality' and 'Go' to numerical value

d = {'UK':0, 'USA':1, 'N':2}
df['Nationality'] = df['Nationality'].map(d)
d = {'YES':1, 'NO':0}
df ['Go'] = df ['Go'].map(d)
print (df)
```

```python
#separate feature column (Predictor) from target column
#x- feature y-target
features = ['Age', 'Experience', 'Rank', 'Nationality']
X = df [features]
y = df['Go']

print (X)
print (y)
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# Instantiate and fit the model
dtree = DecisionTreeClassifier()
dtree.fit(X, y)

# Plot the decision tree
plt.figure(figsize=(12, 8))
plot_tree(dtree, feature_names=features, filled=True)
plt.show()
```

The decision tree uses your earlier decisions to calculate the odds for you to wanting to go see a comedian or not

Rank <= 6.5 means that every comedian with a rank of 6.5 or lower will follow the True arrow (to the left), and the rest will follow the False arrow (to the right).

gini = 0.497 refers to the quality of the split, and is always a number between 0.0 and 0.5, where 0.0 would mean all of the samples got the same result, and 0.5 would mean that the split is done exactly in the middle.

samples = 13 means that there are 13 comedians left at this point in the decision, which is all of them since this is the first step.

value = [6, 7] means that of these 13 comedians, 6 will get a "NO", and 7 will get a "GO".


Decision trees are like the Swiss Army knives of machine learning—simple in form, yet surprisingly powerful.

Interpretability: They’re incredibly easy to understand and visualize. You can trace a path from root to leaf and see exactly how a decision was made, which is great for explaining models to non-technical stakeholders.

Decision trees handle both classification and regression tasks. Whether you're predicting soil type from geotechnical features or estimating groundwater depth, they adapt well.

Handles non-linear relationships: They naturally capture complex interactions between variables without needing you to explicitly define them.

Foundation for ensemble methods: They’re the building blocks for powerful models like Random Forests and Gradient Boosted Trees, which improve accuracy and generalization.
